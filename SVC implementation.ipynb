```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("voice.csv")
```

```python
label=pd.get_dummies(df['label'],drop_first=True)
```

```python
#preprocessing
from sklearn import preprocessing
```

```python
scaler=preprocessing.StandardScaler()
```

```python
scaler.fit(df)
```

```python
X=scaler.transform(df)
```

```python
X.shape
```

```python
Y=label
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=1)
```

```python
from  sklearn.svm import SVC
from sklearn import metrics
```

```python
svc=SVC()
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```

Linear Kernel

```python
svc=SVC(kernel="linear")
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```

Gaussian kernel

```python
svc=SVC(kernel="rbf")
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
svc=SVC(kernel="poly")
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))#overfitting the training set
```

```python
from sklearn.model_selection import cross_val_score
svc=SVC(kernel='linear')
scores=cross_val_score(svc,X,label,cv=10,scoring='accuracy')
print(scores)
print(scores.mean())
```

```python
from sklearn.model_selection import cross_val_score
svc=SVC(kernel='rbf')
scores=cross_val_score(svc,X,label,cv=10,scoring='accuracy')
print(scores)
print(scores.mean())
```

```python
from sklearn.model_selection import cross_val_score
svc=SVC(kernel='poly')
scores=cross_val_score(svc,X,label,cv=10,scoring='accuracy')
print(scores)
print(scores.mean())
```

```python
c_range=list(range(1,26))
acc_score=[]
for c in c_range:
    svc=SVC(kernel='linear',C=c)
    scores=cross_val_score(svc,X,Y,cv=10,scoring='accuracy')
    acc_score.append(scores.mean())
print(acc_score)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(c_range,acc_score)
plt.xticks(np.arange(0,27,2))
plt.xlabel("Value of C for SVC")
plt.ylabel("Cross-Validated score")
plt.show()
```

```python
c_range=list(np.arange(0.1,4,0.1))
acc_score=[]
for c in c_range:
    svc=SVC(kernel='linear',C=c)
    scores=cross_val_score(svc,X,Y,cv=10,scoring='accuracy')
    acc_score.append(scores.mean())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(c_range,acc_score)
plt.xticks(np.arange(0.0,5,0.3))
plt.xlabel("Value of C for SVC")
plt.ylabel("Cross-Validated score")
plt.show()
```

Highest at c=0.1

```python
#taking kernel as rbf and taking different values gamma
```

```python
gamma_range=[0.0001,0.001,0.01,0.1,10,100]
acc_score=[]
for g in gamma_range:
    svc=SVC(kernel='rbf',gamma=g)
    scores=cross_val_score(svc,X,Y,cv=10,scoring='accuracy')
    acc_score.append(scores.mean())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(gamma_range,acc_score)
plt.xticks(np.arange(0.0001,100,5))
plt.xlabel("Value of gamma for SVC")
plt.ylabel("Cross-Validated score")
plt.show()
```

gamma=1 seems to be good

```python
gamma_range=[0.0001,0.001,0.01,0.1]
acc_score=[]
for g in gamma_range:
    svc=SVC(kernel='rbf',gamma=g)
    scores=cross_val_score(svc,X,Y,cv=10,scoring='accuracy')
    acc_score.append(scores.mean())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(gamma_range,acc_score)
plt.xticks(np.arange(0.0001,0.1,0.01))
plt.xlabel("Value of gamma for SVC")
plt.ylabel("Cross-Validated score")
plt.show()
```

gamma=0.01 is best

Polynomial kernel with different degree

```python
degree=[2,3,4,5,6]
acc_score=[]
for d in degree:
    svc=SVC(kernel='poly',degree=d)
    scores=cross_val_score(svc,X,Y,cv=10,scoring='accuracy')
    acc_score.append(scores.mean())
    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(degree,acc_score)
plt.xticks(np.arange(0,6,0.5))
plt.xlabel("degree for svc")
plt.ylabel("Cross-Validated score")
plt.show()
```

degree=3

Grid Search technique to find the best parameter

```python
tuned_params={
    'C':(np.arange(0.1,1,0.1)),'kernel':['linear'],
    'C':(np.arange(0.1,1,0.1)),'gamma':(np.arange(0.01,0.05,0.01)),'kernel':['rbf'],
    'degree':[2,3,4],'gamma':(np.arange(0.01,0.05,0.01)),'C':(np.arange(0.1,1,0.1)),'kernel':['poly']
}
```

```python
from sklearn.model_selection import GridSearchCV
model_svm=GridSearchCV(SVC(),tuned_params,cv=10,scoring='accuracy')
```

```python
model_svm.fit(x_train,y_train)
```

```python
model_svm.best_score_
```

```python
model_svm.best_params_
```

```python
y_pred=model_svm.predict(x_test)
print(metrics.accuracy_score(y_pred,y_test))
```
